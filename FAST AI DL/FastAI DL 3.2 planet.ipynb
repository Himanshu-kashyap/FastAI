```
# f2 is a function defined by the contest to measure how good our function is. You can always define your own evaluation function.

from planet import f2

metrics=[f2]
f_model = resnet34
```

```
label_csv = f'{PATH}train_v2.csv'
data_size = len(list(open(label_csv))) -1

# Generate some data to use as validation data by taking some random rows (20%)
val_idxs = get_cv_idxs(data_size)
```

```
def get_data(image_size):
  transforms = tfms_from_model(f_model,image_size,aug_tfms = transform_top_down,max_zoom =1.05)
  return ImageClassifierData.from_csv(PATH,'train-jpg',transforms = transforms,suffix =".jpg",
                                     val_idxs = val_idxs,test_name ='test-jpg')

# ImageDataClassifier : path, training_set_path, transforms, suffix, val set, test_set_path :: type = dATA
  
```

```
data = get_data(256)
```
